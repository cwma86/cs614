## Stock Website Recommender

## Pitch: 
In a previous project I develop a Stock viewing website (https://mauer-stocks.azurewebsites.net/), and in this project I implemented a recommender system that can recommend stocks for users. The goal of this system is to improve the user experience while using the site and allow users to easily find new stocks. 

## Data source:
Since I wanted to develop a system that worked well with my website I was unable to leverage an existing data set and instead leveraged the favorites data base used by my website in conjuction with this stock listing website (https://companiesmarketcap.com/all-categories/). My data set was created by creating various users with randomly selected stocks accross various categories such as 
* Tech
* Retail 
* Realestate
* Finance
* Defense
* Health
As well as some mixed category users.  The idea behind the creation of this data set is that I expect many users to focus primarily on specific types of stocks, and wanted a system to would recommend similar stocks to users. 

## Model and data justification:
For model selection I chose to go with collaborative filtering via matrix factorization approach using Non-negative matrix factorization (sklearn.decomposition.NMF). Since my data set was focused on a binary recommender (user has/has not favorited) and I wanted a highly performant system that could be updated everytime a new user is added, or stock is favorited this seemed like an optimal choice. Since I am currently using a fairly small data set (24 users and 253 stocks) spread across 7 independant types, I decided to perform factorizations using 10 features

## Commented examples:
The input of the model uses a sparse matrix built from users (rows) and their favorited stocks (columns). 

```python
import os
import pandas as pd
from colab_filter import get_users, get_stocks, build_truth_matrix

truth_matrix = build_truth_matrix(get_users(), get_stocks())
pd.options.display.width = 150
pd.set_option('display.max_columns', 20)
pd.set_option('display.max_rows', 20)

## example of the sparse matrix that includes all user favorites
print(truth_matrix)
```

```python
# example of the factored user matrix with each users trained weights
from colab_filter import train_weights

number_of_features = 10 # number of features to be factorized
user_weights, stock_weights = train_weights(truth_matrix, number_of_features)
user_weights = pd.DataFrame(user_weights, index=get_users())
print(user_weights)
```

```python
# example of the factored stock matrix with each stocks trained weights
stock_weights = pd.DataFrame(stock_weights, columns=get_stocks())
pd.set_option('display.max_columns', 10)
print(stock_weights)
```

Note that we are able to factor our truth matrix down to 2,530 parameters and 240 parameters down from the prediction matrix size of 6,0072 parameters. This allows us to calculate user favorites by multiplying the users row with the stock matrix which results in a 1x240 matrix(array) which can easily be sorted to find the users highest recommended stocks. 

Note: That you can kind of see how the factorized matrices are gouping stocks toghether, for example AAPL (Apple) has high rating for feature 0, similar to the tech stock users.  

## Testing:
Since this recomender system is meant to provide real world results and help users find useful stocks there is no Truth data per say, and therefore validation metrics are fairly subjective. However here are some example of stocks being recommended for various users. 

```python
from colab_filter import recommend_stocks
user_weights_path = "./pred_user_weights"
stock_weights_path = "./pred_stock_weights"

user = 'tech2'
recommended_stocks = recommend_stocks(user, user_weights_path, stock_weights_path)
print(f"reccomended stocks for {user} \n{recommended_stocks}")

user = 'retail'
recommended_stocks = recommend_stocks(user, user_weights_path, stock_weights_path)
print(f"reccomended stocks for {user} \n{recommended_stocks}")

user = 'realestate'
recommended_stocks = recommend_stocks(user, user_weights_path, stock_weights_path)
print(f"reccomended stocks for {user} \n{recommended_stocks}")

user = 'techfinance'
recommended_stocks = recommend_stocks(user, user_weights_path, stock_weights_path)
print(f"reccomended stocks for {user} \n{recommended_stocks}")
```

of the 4 provided examples you can see `tech2` does a good job of providing all tech stock with the exception of `BRK.B` which arguably belongs more in finance catagory.

Additionally you can see how `techfinance` is able to recommend the user from stocks in both tech and finance catagories, without having any other similar users that have favorited both in combination.
